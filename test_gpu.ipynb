```python
import numpy as np
import matplotlib.pyplot as plt

# Simulating the steps (time) data for 6 million steps
steps_million = np.linspace(0, 6_000_000, num=300)

# Creating placeholder data for Alinea Agent's average episode length for plotting
episode_length_alinea = np.concatenate([np.linspace(40000, 65000, num=150), np.linspace(65000, 55000, num=150)])

# Plot the Alinea Agent's average episode length without "Modified" in the title
plt.figure(figsize=(10, 6))
plt.plot(steps_million, episode_length_alinea, label='Alinea Agent', color='orange', linestyle='-', marker='d')

# Add title and labels, removing the "Modified" label
plt.title('Average Episode Length: Alinea Agent')
plt.xlabel('Steps (in Millions)')
plt.ylabel('Average Episode Length')
plt.legend()

# Display the plot
plt.show()
```
