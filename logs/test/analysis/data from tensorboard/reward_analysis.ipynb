```python
import pandas as pd 
```

```python
data_path = "/Users/cheimamezdour/Projects/PFE/DQN-ITSCwPD/logs/test/analysis/data from tensorboard/J7_TLS_DuelingDoubleDQNAgent_lr0.0001-AvgRew.csv"
data = pd.read_csv(data_path)
```

```python
import matplotlib.pyplot as plt

# Plot Value per Step
plt.figure(figsize=(10, 6))
plt.plot(data['Step'], data['Value'], marker='o', linestyle='-')
plt.title('Average Reward per Step')
plt.xlabel('Step')
plt.ylabel('Average Reward')
plt.grid(True)
plt.show()
```

## After smoothing 

```python
# Calculate the exponential moving average (EMA) for smoothing
smoothing_factor = 0.99
smoothed_values = data['Value'].ewm(alpha=1-smoothing_factor).mean()

# Plot the original and smoothed Value per Step as a line plot
plt.figure(figsize=(10, 6))
plt.plot(data['Step'], data['Value'], linestyle='-', alpha=0.5, label='Original')
plt.plot(data['Step'], smoothed_values, linestyle='-', label='Smoothed (0.99)')
plt.title('Average Reward per Step with Smoothing')
plt.xlabel('Step')
plt.ylabel('Average Reward')
plt.legend()
plt.grid(True)
plt.show()
```
