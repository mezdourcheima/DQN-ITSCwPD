# Analysis of Model Performance

## Step 1: Statistical Summary

```python

import pandas as pd

# Load the CSV file
file_path = '/Users/cheimamezdour/Projects/PFE/DQN-ITSCwPD/logs/test/J7_TLS/con/dataDuelingDoubleDQNAgent_lr0.0001.csv'
data = pd.read_csv(file_path)

# Statistical summary of the dataset
stat_summary = data.describe()
stat_summary
```

```python

# Correlation matrix for numerical columns
numeric_columns = data.select_dtypes(include='number').columns
correlation_matrix = data[numeric_columns].corr()
correlation_matrix
```

## Step 3: Trend Analysis

```python

import matplotlib.pyplot as plt

# Trend analysis over episodes
fig, axs = plt.subplots(4, 1, figsize=(12, 24))

# Rewards trend
axs[0].plot(data['ep'], data['r'], label='Rewards')
axs[0].set_xlabel('Episode')
axs[0].set_ylabel('Rewards')
axs[0].set_title('Rewards over Episodes')
axs[0].legend()

# Sum Delay trend
axs[1].plot(data['ep'], data['sum_delay'], label='Sum Delay', color='orange')
axs[1].set_xlabel('Episode')
axs[1].set_ylabel('Sum Delay')
axs[1].set_title('Sum Delay over Episodes')
axs[1].legend()

# Sum Waiting Time trend
axs[2].plot(data['ep'], data['sum_waiting_time'], label='Sum Waiting Time', color='green')
axs[2].set_xlabel('Episode')
axs[2].set_ylabel('Sum Waiting Time')
axs[2].set_title('Sum Waiting Time over Episodes')
axs[2].legend()

# Sum Queue Length trend
axs[3].plot(data['ep'], data['sum_queue_length'], label='Sum Queue Length', color='red')
axs[3].set_xlabel('Episode')
axs[3].set_ylabel('Sum Queue Length')
axs[3].set_title('Sum Queue Length over Episodes')
axs[3].legend()

plt.tight_layout()
plt.show()
```

#### For 1000 steps

```python
import numpy as np

# Simulating data for 1000 episodes/steps
# Assuming the data has columns similar to those mentioned in the plots: 'ep', 'r', 'sum_delay', 'sum_waiting_time', 'sum_queue_length'

np.random.seed(42)  # For reproducibility

# Generate 1000 episodes
steps = 1000
data = {
    'ep': np.arange(1, steps + 1),
    'r': np.random.normal(0, 1, steps).cumsum(),  # Cumulative reward, for instance
    'sum_delay': np.random.uniform(0, 10, steps).cumsum(),  # Cumulative sum of delays
    'sum_waiting_time': np.random.uniform(0, 5, steps).cumsum(),  # Cumulative sum of waiting time
    'sum_queue_length': np.random.uniform(0, 3, steps).cumsum(),  # Cumulative sum of queue length
    'avg_waiting_time': np.random.uniform(1, 3, steps),  # Average waiting time per episode
    'avg_queue_length': np.random.uniform(1, 5, steps),  # Average queue length per episode
    'total_density': np.random.uniform(10, 20, steps),  # Total density per episode
    'total_flow': np.random.uniform(5, 15, steps)  # Total flow per episode
}

# Convert the data dictionary to a DataFrame
data_df = pd.DataFrame(data)

import matplotlib.pyplot as plt

# Plotting the trends over 1000 episodes

fig, axs = plt.subplots(4, 1, figsize=(12, 24))

# Rewards trend
axs[0].plot(data_df['ep'], data_df['r'], label='Rewards')
axs[0].set_xlabel('Episode')
axs[0].set_ylabel('Rewards')
axs[0].set_title('Rewards over Episodes')
axs[0].legend()

# Sum Delay trend
axs[1].plot(data_df['ep'], data_df['sum_delay'], label='Sum Delay', color='orange')
axs[1].set_xlabel('Episode')
axs[1].set_ylabel('Sum Delay')
axs[1].set_title('Sum Delay over Episodes')
axs[1].legend()

# Sum Waiting Time trend
axs[2].plot(data_df['ep'], data_df['sum_waiting_time'], label='Sum Waiting Time', color='green')
axs[2].set_xlabel('Episode')
axs[2].set_ylabel('Sum Waiting Time')
axs[2].set_title('Sum Waiting Time over Episodes')
axs[2].legend()

# Sum Queue Length trend
axs[3].plot(data_df['ep'], data_df['sum_queue_length'], label='Sum Queue Length', color='red')
axs[3].set_xlabel('Episode')
axs[3].set_ylabel('Sum Queue Length')
axs[3].set_title('Sum Queue Length over Episodes')
axs[3].legend()

plt.tight_layout()
plt.show()

# Additional visualizations for key metrics
fig, axs = plt.subplots(4, 1, figsize=(12, 24))

# Average Waiting Time
axs[0].plot(data_df['ep'], data_df['avg_waiting_time'], label='Average Waiting Time', color='purple')
axs[0].set_xlabel('Episode')
axs[0].set_ylabel('Average Waiting Time')
axs[0].set_title('Average Waiting Time over Episodes')
axs[0].legend()

# Average Queue Length
axs[1].plot(data_df['ep'], data_df['avg_queue_length'], label='Average Queue Length', color='brown')
axs[1].set_xlabel('Episode')
axs[1].set_ylabel('Average Queue Length')
axs[1].set_title('Average Queue Length over Episodes')
axs[1].legend()

# Total Density
axs[2].plot(data_df['ep'], data_df['total_density'], label='Total Density', color='blue')
axs[2].set_xlabel('Episode')
axs[2].set_ylabel('Total Density')
axs[2].set_title('Total Density over Episodes')
axs[2].legend()

# Total Flow
axs[3].plot(data_df['ep'], data_df['total_flow'], label='Total Flow', color='magenta')
axs[3].set_xlabel('Episode')
axs[3].set_ylabel('Total Flow')
axs[3].set_title('Total Flow over Episodes')
axs[3].legend()

plt.tight_layout()
plt.show()
```
